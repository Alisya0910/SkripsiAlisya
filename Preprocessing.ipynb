#Preprocessing data Ganjar Pranowo





```python
tweet_ganjar = pd.read_excel('GanjarPresidentTweets.xlsx')
tweet_ganjar
```

```python
#membuat attribut hashtag dengan nilai hashtag dari attribut full_text
tweet_ganjar['hashtag'] = tweet_ganjar['full_text'].apply(lambda x: re.findall(r"#(\w+)",x))
tweet_ganjar
```

Membersihkan text dari URL, Mention, dll

```python
#Text-Cleaning (URLs, Mentions, etc)
#installing tweet-preprocessor
!pip install tweet-preprocessor
```

```python
import preprocessor as p
```

```python
def preprocess_tweet(row):
    text = row['text']
    text = p.clean(text)
    return text
```

```python
#mengimplementasikan fungsi membersihkan attribut full_name ke dalam data
tweet_ganjar['text'] = tweet_ganjar.apply(preprocess_tweet, axis=1)
```

```python
#data setelah diberihkan dari url,mentions, dll
tweet_ganjar
```

Proses Tekonization, removal digits, stop words and punctuations or proses normal text preprocessing (lowercasing, punctuation removal, replace extra white spaces, and stopwords removal)

```python
from gensim.parsing.preprocessing import remove_stopwords
```

```python
def stopword_removal(row):
    text = row['text']
    text = remove_stopwords(text)
    return text
```

```python
tweet_ganjar['full_text'] = tweet_ganjar.apply(stopword_removal, axis=1)
```

```python
tweet_ganjar
```

```python
#menghapus extra white space, punctuation, dan menlowercase kan
tweet_ganjar['full_text'] = tweet_ganjar['full_text'].str.lower().str.replace('[^\w\s]',' ').str.replace('\s\s+', ' ')
```

```python
tweet_ganjar
```
